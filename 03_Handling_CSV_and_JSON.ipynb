## 📊 Topic 1: Handling CSV Files
CSV files are just text files where columns are separated by commas.
We *could* read them using `.split(",")`, but Python has a built-in `csv` library that handles the hard work (like ignoring commas inside quotes).

### Reading a CSV

#### Using `csv.reader()`
The `csv.reader()` function reads a CSV file and returns an iterable of rows, where each row is a list of values.

````python
import csv

print("--- Reading CSV ---")
with open(os.path.join("resources", "students.csv"), "r", newline="") as file:
    reader = csv.reader(file) # Create a reader machine
    
    # Loop through the reader to get rows
    for row in reader:
        print(row) # Each row is a List of Strings
````

#### Using `csv.DictReader()`
The `csv.DictReader()` function reads a CSV file and returns an iterable of rows, where each row is a dictionary with column headers as keys.

````python
import csv

print("--- Reading CSV ---")
with open(os.path.join("resources", "students.csv"), "r", newline="") as file:
    reader = csv.DictReader(file) # Create a DictReader machine
    
    # Loop through the reader to get rows
    for row in reader:
        print(row) # Each row is a dictonary with key as coulumn name and values as Strings
````

### CSV is dumb (Strings only)

Gotcha: When you read from CSV, `10` becomes `"10"` (String).

Comparison: In JSON, `10` stays `10` (Number).

Remember in Level 2 problems's solution (`row[1]` is a string, so we didn't do math on it).

### Writing/Appending to CSV
Just like text files, we use mode `"w"` to write and `"a"` to append.

#### Writing with `csv.writer()`
The `csv.writer()` function allows you to write rows to a CSV file. 

You can use the `writerow()` method to write a single row or `writerows()` to write multiple rows at once.

````python
# using writerow
new_student = ["Bruce", "17", "A"]

with open(os.path.join("resources","students.csv"), "a", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(new_student)

print("Added Bruce to the CSV.")
````

````python
# Writing Multiple Rows using csv.writer()
new_students = [["Bruce", "17", "A"],
                ["Clint", "16", "B+"],
                ["Wanda", "15", "A-"],]
# using writerow
with open(os.path.join("resources","students.csv"), "a", newline="") as f:
    writer = csv.writer(f)
    for student in new_students:
        writer.writerow(student)
print("Added multiple students to the CSV.")
````

````python
# writing mulitple row using writerows
new_students = [["Vision", "16", "A"],
                ["Sam", "17", "B"],
                ["Bucky", "16", "B-"],]

with open(os.path.join("resources","students.csv"), "a", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(new_students)
print("Added multiple students to the CSV using writerows.")
````

#### Writing with `csv.DictWriter()`
The `csv.DictWriter()` function allows you to write rows to a CSV file using dictionaries. You need to specify the fieldnames (column headers) when creating the writer object.

You can use the `writerow()` method to write a single row or `writerows()` to write multiple rows at once.

When creating a new CSV file, calling `writeheader()` is recommended to write the column headers to the CSV file. If you're appending to an existing file, you should not call `writeheader()` again, as it will add the headers in the middle of your data.

````python
# Creating a new file and writing multiple rows using writerow
with open(os.path.join("resources","new_students.csv"), "w", newline="") as f:
    fieldnames = ["Name", "Age", "Grade"]
    writer = csv.DictWriter(f, fieldnames=fieldnames)

    writer.writeheader()   # RECOMMENDED to write the header when creating a new file
    writer.writerow({"Name": "Alice", "Age": 23, "Grade": "A"})
    writer.writerow({"Name": "Bob", "Age": 25, "Grade": "B"})
    writer.writerow({"Name": "Charlie", "Age": 22, "Grade": "A-"})

print("Created 'new_students.csv' and added students using DictWriter.")
````

````python
# writing mulitple row using writerow

new_students_dict = [
    {"Name": "David", "Age": 24, "Grade": "B+"},
    {"Name": "Eve", "Age": 23, "Grade": "A"},
    {"Name": "Frank", "Age": 25, "Grade": "B-"},]

with open(os.path.join("resources","new_students.csv"), "a", newline="") as f:
    fieldnames = ["Name", "Age", "Grade"]
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    
    # No need to write header when appending to an existing file
    writer.writerows(new_students_dict)
````

### The `newline=""` Requirement

The Bug: If you run CSV write code on Windows without `newline=""`, you will get empty blank lines between every row.

The Fix: Always open file with newline="" like `open(..., newline="")` when writing CSVs. It's a standard boilerplate.

## 🕸️ Topic 2: Handling JSON Files
JSON looks **exactly** like a Python Dictionary.
It is the standard format for web data (Twitter API, Weather apps, etc.).

| Python | JSON |
| :--- | :--- |
| Dictionary | Object `{}` |
| List | Array `[]` |
| True/False | true/false |
| None | null |

### Example of JSON

```json
{
  "firstName": "John",
  "lastName": "Doe",
  "age": 30,
  "isStudent": false,
  "address": {
    "street": "123 Maple St",
    "city": "Anytown",
    "zip": "12345"
  },
  "courses": ["History", "Art", "Biology"]
}
```

We use the `json` library.
* `json.load(file)`: Read JSON file $\rightarrow$ Python Dict.
* `json.dump(data, file)`: Python Dict $\rightarrow$ Write JSON file.

````python
import json

# 1. Reading Existing JSON file 
print("--- Reading JSON ---")
with open(os.path.join("resources","school_config.json"), "r", newline="") as file:
    school_data = json.load(file) # Converts text -> Dictionary

# extract and print some data
print(f"School Name: {school_data['school']}")
print(f"Student 1: {school_data['students'][0]}")
````

````python
# Modifying & Saving JSON
with open(os.path.join("resources","school_config.json"), "r", newline="") as file:
    school_data = json.load(file) # Converts text -> Dictionary
# As dictions are mutable, we can modify the data in place
school_data["principal"] = "Mr. Morita" # Add new key

# If you want to update the existing JSON file, you need to write the modified data back to the file. This will overwrite the existing content with the new data.
# Note: If you want to keep the original file and create a new file with the updated data, you can use a different filename in the `open()` function.
with open(os.path.join("resources","school_config.json"), "w", newline="") as file:
    json.dump(school_data, file, indent=4) # Save it back

print("JSON updated.")
````

### `dump` vs `dumps`

Confusion: Students will google and find `json.dumps()` (with an 's').

Clarification:

`dump` = File (Dump it into the hard drive).

`dumps` = String (Dump it into a String variable).

````python
# We already now how dump works, let's see how dumps works
# Using `dumps` to convert to a JSON string
# code to show different between `dump` and `dumps`
import json
data = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}
json_string = json.dumps(data, indent=4)
print("Data converted to JSON string using json.dumps():")
print(json_string)
print("Type of data:", type(data))
print("Type of json_string:", type(json_string))
````

### `load` vs `loads`
Confusion: Students will google and find `json.loads()` (with an 's').

Clarification:

`load` = File (Load it from the hard drive).

`loads` = String (Load it from a String variable).


````python
# code to show different between `load` and `loads`
# We already know how load works, let's see how loads works
json_string = '{"name": "Alice", "age": 30, "city": "New York"}'
# Using `loads` to convert JSON string to Python object
data_from_string = json.loads(json_string)
print("Data loaded from JSON string using json.loads():")
print(data_from_string)
print("Type of json_string:", type(json_string))
print("Type of data_from_string:", type(data_from_string))
````

### Chaining Dumps Will Break Your Code
Calling `json.dumps()` multiple times on the same data will convert it to a JSON string, then convert that string to a JSON string, and so on. This can lead to unexpected results if you call dumps multiple times on the same data without realizing it. 
Same is applicable for `json.loads()`.

````python
data = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}
json_string = json.dumps(data, indent=4)
json_string_again = json.dumps(json_string, indent=4)
json_string_load = json.loads(json_string_again)
json_string_load_again = json.loads(json_string_load)
print("Original data:", data)
print("After first dumps:", json_string)
print("After second dumps:", json_string_again)
print(json_string_load_again)
# see how \n and spaces were added in the second dumps, because it treated the first json_string as a regular string and added escape characters for the newlines and spaces.
````
