## CSV & JSON
- Structured data
    - CSV (Comma Separated Values): Simple, row-column data (like Excel)
    - JSON (JavaScript Object Notation): Nested key-value data (like Dictionaries)

### Working with csv:
- Always open file csv with `newline = ""` like `open(..., newline="")` when writing CSVs
- **Reading a csv**:

**Using Reader:**
```Python
    with open(os.path.join("resources", "students.csv"), "r", newline="") as file:
        reader = csv.reader(file)
```

**Using DictReader:**
```Python
    with open(os.path.join("resources", "students.csv"), "r", newline="") as file:
        reader = csv.DictReader(file)
```
- **Writing/Appending to csv**:

**Using Writer:**
```Python
    with open(os.path.join("resources","students.csv"), "a", newline="") as file:
        writer = csv.writer(file)
```

**Using DictWriter:**
```Python
    with open(os.path.join("resources","students.csv"), "a", newline="") as file:
        writer = csv.DictWriter(file, fieldnames=["Name", "Age", "Grade"])
```
- Methods in csv:
    - `reader()` - to read CSV data, Returns a reader object which will iterate over lines in the given CSV file.
    - `writer()` 
        - To write CSV data, Returns a writer object responsible for converting the user's data into delimited strings on the given file-like object.
        - **To Write Header**: To write header row to CSV file, you need to call `writerow()` method as first call then call `writerow()` method for each row of data.
        - **Write single row using `writerow()`**: To write a row to the CSV file, pass a list of values to the `writerow()` method on the writer object. Length of the list should match the number of columns/headers in the CSV file.
        - **Multiple rows**: To write multiple rows at once, pass a list of lists to the `writerows()` method on the writer object.
        - **Note: **Make sure to use `newline=""` when opening the file for writing to avoid adding extra blank lines in the CSV file.
    - `DictReader()` 
        - To read CSV data into a list of dictionary, where the keys are taken from the first row of the CSV file.
    - `DictWriter()` 
        - To write CSV data from dictionaries, where the keys are used as fieldnames in the CSV file. Fieldnames must be provided when creating the DictWriter object. This prevents errors if the dictionary keys are in different order or if some keys are missing.
        - **Write header**: To write header row (fieldnames) to the CSV file you have to call the `writeheader()` method on the DictWriter object.
        - **Write single row**: To write data, pass dictionary to the `writerow()` method on the DictWriter object.
        - **Write multiple rows**: To write multiple rows at once, pass a list of dictionaries to the `writerows()` method on the DictWriter object.


### Working with JSON files:
- `Json.dump(data, file)`: Python Dict --> Write JSON file
- `Json.dumps(data)`: Python Dict --> Write JSON string
- `Dump` vs `dumps`:
    - `Dump` = File (Dump it into the hard drive).
    - `Dumps` = String (Dump it into a String variable).
- `Json.load(file)`: Read JSON file --> Python Dict.
- `Json.loads(json_string)`: Read JSON string --> Python Dict.
- `load` vs `loads`
- `load` = read JSON encoded data from a file and convert it into a Python dictionary
- `loads` = parse a valid JSON string and convert it into a Python Dictionary
