## 📡 Topic 2: How the Web Actually Works (HTTP Basics)

Every time a browser loads a website or a Python script calls an API, the same thing happens behind the scenes.

A **client** sends an **HTTP request** to a **server**, and the server sends back an **HTTP response**.

Example:

Your Python script asks a weather API for today's temperature.

1. Your code sends a **request**
2. The server processes it
3. The server sends back a **response**

This simple request-response cycle powers the entire web.

### 📬 HTTP Request Structure

An HTTP request has four main component when making a call from client to the server.
1. **Method**:
    - The method tells the server what action the client wants to perform.
    - Example: `GET https://api.example.com/users`
2. **URL**:
    - The URL tells the server which resource we want to interact with.
    - Example: `https://api.example.com/users`
3. **Headers**: 
    - Headers provide additional information about the request.
    - Example: `Authorization: Bearer token123` or `Content-Type: application/json`
4. **Body**
    - The body contains the data being sent to the server.
    - Example: `{"name": "Alice", "email": "alice@example.com"}`

### 📬 HTTP Response Structure
After receiving a request, the server sends back a response. Nearly every API uses JSON.

A response typically contains:

- Status code
- Headers
- Body

Example: 
```python
{
 "name": "Alice",
 "email": "alice@example.com",
 "id": 101
}
```

### 🚦 HTTP Status Codes
Status codes tell the client what happened to the request. They are three-digit numbers.

Instead of reading long error messages, developers usually look at the status code first.

#### 200 Range → Success

| Code | Meaning                |
| ---- | ---------------------- |
| 200  | Request successful     |
| 201  | Resource created       |
| 204  | Success but no content |


#### 400 Range → Client Errors
These happen when the request sent by the client is incorrect.

| Code | Meaning            |
| ---- | ------------------ |
| 400  | Bad request        |
| 401  | Unauthorized       |
| 403  | Forbidden          |
| 404  | Resource not found |


#### 500 Range → Server Errors
These indicate problems on the server side.

| Code | Meaning               |
| ---- | --------------------- |
| 500  | Internal server error |
| 502  | Bad gateway           |
| 503  | Service unavailable   |


### 🔄 HTTP Methods
HTTP methods define what action the client wants to perform.

| Method | Purpose          |
| ------ | ---------------- |
| GET    | Retrieve data    |
| POST   | Create something |
| PUT    | Replace data     |
| PATCH  | Modify data      |
| DELETE | Remove data      |

## 🔍 Topic 5: Query Parameters (`?key=value`)

Sometimes we don't want all the data; we want to search for something specific.
We do this using Query Parameters. In a URL, this looks like `?search=pizza&limit=5`. Any thing added post `?` in a url is query parameter.

Instead of typing that manually into the string, the `requests` library lets us pass a dictionary of `params`.

````python
base_url = "https://api.github.com/users"

# We want to search for users whose id is greater than 42804162
search_params = {
    "since": "42804162"
}

# Pass the dictionary into the 'params' argument
response = requests.get(base_url, params=search_params)

# Python automatically builds this URL: 
print(f"URL created: {response.url}")

data = response.json()

# The API returns a list of dictionaries. Let's print the first 3.
print("\nTop 3 results:")
for user in data[:3]:
    print(f"- {user['login']}")
````

## ⚠️ Topic 6: Error Handling

You should never write:

````python
url = "https://api.github.com/users"
response = requests.get(url)
data = response.json()
````

**NEVER TRUST THE ENDUSER**. Always do defensive coding.

````python
# Method 1: using explicit status code check
url = "https://api.github.com/users"
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
else:
    print("API error:", response.status_code)
````

````python
# Method 2. using raise_for_status

url = "https://api.github.com/usersd"
response = requests.get(url)
response.raise_for_status() # Will raise an exception if the status code is not 200
data = response.json()
````

````python
# Method 2. using raise_for_status

url = "https://api.github.com/users"
response = requests.get(url)
if response.ok: # Will raise an exception if the status code is not 200
    print(response.json())
else:
    print("Error Occured:", response.status_code)
````

## ⏱️ Topic 8: Timeouts and Reliability

Network requests can hang indefinitely if the server does not respond.

To prevent this, requests should include a timeout.

````python
url = "https://api.github.com/users"
requests.get(url, timeout=5)
````

**Meaning**:

Wait up to 5 seconds before raising an error.

This prevents programs from freezing due to network issues.

## 🚦 Topic 9: Rate Limiting

Many APIs limit how many requests a client can make.

Reasons include:
- Preventing abuse
- Protecting server resources
- Ensuring fair usage

Example limit:
```bash
100 requests per hour
```

If exceeded, the API may return:
```bash
429 Too Many Requests
```

Developers must sometimes implement retry logic or wait before making more requests.

## 📚 Topic 11: Pagination

Large APIs never return everything at once.

Example:

/users?page=1

/users?page=2

You have to iterate pages to get all data.

````python
import requests
import json

url = "https://api.github.com/repos/geekychandraul/chat-with-your-docs/issues"

params = {"per_page": 4, "page": 1}

headers = {
    "Authorization": f"Bearer {github_access_token}"
}
response = requests.get(url, headers=headers, params=params)
response.raise_for_status()
for res in response.json():
    print(f"{res["number"]}: summary: {res['title']}: state: {res['state']}")
````

````python
import requests
import json

url = "https://api.github.com/repos/geekychandraul/chat-with-your-docs/issues"

params = {"per_page": 2, "page": 2}

headers = {
    "Authorization": f"Bearer {github_access_token}"
}
response = requests.get(url, headers=headers, params=params)
response.raise_for_status()
for res in response.json():
    print(f"{res["number"]}: summary: {res['title']}: state: {res['state']}")
````

````python
limit = 20
offset = 0
is_countine = True
page = 0
# per_page = 10
url = "https://pokeapi.co/api/v2/pokemon"
pokemon_list = []
while is_countine:
    # params = {"page": page}
    params = {"limit": limit, "offset": offset}
    response = requests.get(url, params=params)
    response.raise_for_status()
    data = response.json()

    for pokemon in data["results"]:
        pokemon_list.append(pokemon["name"])

    if data["next"] and page < 5:
        offset += limit
        page += 1
        is_countine = True
    else:
        is_countine = False

print(pokemon_list)
````

## 🔍 Topic 12: Sending Data (POST, PUT, PATCH, DELETE Requests)

You can also send data to server by sending as JSON in the HTTP Request. It is also called as `reqeust body`.

### Post HTTP Method

````python
import requests
from datetime import datetime
url = "https://api.github.com/repos/geekychandraul/chat-with-your-docs/issues"

headers = {
    "Authorization": f"Bearer {github_access_token}"
}

data = {
    "title": f"{datetime.now().strftime('%Y-%m-%d %H:%M:%S')} Created using API call",
    "body": "Created via API"
}

response = requests.post(url, headers=headers, json=data)

print(response.json())
issue_number = response.json()["number"]
````

### PUT HTTP METHOD

````python
# Update an issue
import requests

url = f"https://api.github.com/repos/geekychandraul/chat-with-your-docs/issues/{issue_number}"

headers = {
    "Authorization": f"Bearer {github_access_token}"
}

data = {
    "title": "Updated: API Demo Issue"
}

response = requests.patch(url, headers=headers, json=data)

print(response.json())
````

### PATCH HTTP METHOD

````python
# Update an issue
import requests

url = f"https://api.github.com/repos/geekychandraul/chat-with-your-docs/issues/{issue_number}"

headers = {"Authorization": f"Bearer {github_access_token}"}

data = {"state": "closed"}

response = requests.patch(url, headers=headers, json=data)

print(response.json())
````

### 📦 Request Body vs 🔎 Query Parameters

The key difference is where the data is sent and what it is used for.

| Feature            | Query Parameters                 | Request Body               |
| ------------------ | -------------------------------- | -------------------------- |
| Where data appears | In the URL                       | Inside the request payload |
| Common methods     | GET                              | POST, PUT, PATCH           |
| Purpose            | Filtering or modifying retrieval | Sending actual data        |
| Visibility         | Visible in the URL               | Not visible in the URL     |

Query Param: 
https://api.example.com/products?user_name=abc&password=1234 ==> Visible to user

Unlike query parameters, the data or request body is not part of the URL.

NOTE: Avoid sensitive data in query parameters.


### 🤠 `data` vs `json` Parameter in Requests
The `data` and `json` parameters are used to send data in the body of a request, but they handle the data differently: 
`data` Parameter 
- **Purpose**: Primarily used for sending form-encoded data (`Content-Type: application/x-www-form-urlencoded`). 
- **Input Types**: Accepts a dictionary, a list of tuples, bytes, or a file-like object. 
- **Automatic Behavior**: 
	- If you pass a dictionary, Requests automatically form-encodes it. 
	- If you pass a string or bytes, Requests sends it as is, and you might need to set the `Content-Type` header manually.

`json` Parameter:
- **Purpose**: Specifically designed for sending JSON-encoded data (`Content-Type: application/json`). 
- **Input Types**: Accepts a Python dictionary (or other JSON-serializable object). 
- **Automatic Behavior**: Requests automatically serializes the Python object to a JSON string using Python's built-in `json` module and sets the ontent-Type header to `application/json` for you.
