## ⚡ Topic 1: Lambda Functions
A **Lambda** is a small, anonymous function.
* **Old Way (`def`):**
    ```python
    def square(x):
        return x * x
    ```
* **Lambda Way:**
    ```python
    square = lambda x: x * x
    ```

**Syntax:** `lambda arguments : expression`

````python
def square(x):
    return x * x    
temp_sq = square

print(temp_sq(5))
print(temp_sq(10))

# conveting to lambda
square = lambda x: x * x

print(square(5))
print(square(10))
````

````python
# A simple lambda to double a number
doubler = lambda x: x * 2

print(doubler(5))
print(doubler(10))

# Lambda with multiple inputs
multiply = lambda a, b: a * b

print(multiply(5, 6))
````

### ⚠️ When to use them?
Don't replace every function with a lambda!
Use them when you need a short function for a **short period of time** (usually inside another function).


### Lambda vs Def

Best Practice: "Use `def` for logic you use twice. Use `lambda` for logic you use once (inside a map/filter/sort)."

## 🗺️ Topic 2: Map and Filter
These are built-in functions that process lists. They are faster and cleaner than writing `for` loops.



### 1. `map(function, list)`
Apply a function to **every** item in a list.

````python
prices = [10, 20, 30, 40]

# Goal: Add $5 tax to every price
# Option A: The Loop Way (Boring)
# Option B: The Map Way (Cool)

updated_prices = list(map(lambda x: x + 5, prices))
updated_prices = list(map(lambda x: x + 5, [10, 20, 30, 40]))


print(f"Original: {prices}")
print(f"Updated:  {updated_prices}")
````

````python
# Example with normal python fucntion (without using lambda)
prices = [10, 20, 30, 40]

def add_five(price):
    # Here you can add complex logic also which you can not do in lambda.
    return price + 5

updated_prices = list(map(add_five, prices))
print(updated_prices)
````

### 2. `filter(function, list)`
Keep only the items where the function returns `True`. If the fucntion returns anything which evaluates to False, value will be filtered out.

Remember False Values in Python : `False`, `None`, `0`, `0.0`, `0j`, `''`, `[]`, `()`, `{}`, `set()`, `range(0)`

````python
ages = [5, 12, 17, 18, 24, 32]

# Goal: Get only adults (18+)
adults = list(filter(lambda x: x >= 18, ages))

print(f"Adults: {adults}")
````

````python
ages = [5, 12, 17, 18, 24, 32]

# Example with normal python fucntion (without using lambda)
def filter_by_age(age):
    if age >= 18:
        return True

# Goal: Get only adults (18+)
adults = list(filter(filter_by_age, ages))

print(f"Adults: {adults}")
````

````python
ages = [5, 12, 17, 18, 24, 32]

# Example with normal python fucntion (without using lambda)
def filter_by_age(age):
    if age >= 18:
        return "Able To Vote" # string with value is considered True

# Goal: Get only adults (18+)
adults = list(filter(filter_by_age, ages))

print(f"Adults: {adults}")
````

### `list()` Casting is Mandatory

The Gotcha: In Python 3, `map()` and `filter()` return "Iterators" (objects waiting to run), not Lists.

The Symptom: If students print `map(...)`, they get `<map object at 0x7f...>` instead of `[15, 25, 35, 45]`.

The Fix: Always wrap them: `list(map(...))`.

### Using Lambda function is not Mandatory. 
Using lambda as a fuciton in map, filter or reduce is not mandatory. You can also use normal python fucntions also provided they return the expected output.

## 🎒 Topic 3: `reduce(function, list, initial_value)`
Sometimes you would like to reduce a list into a single value.
reduce is not a built-in fucntion. You need to import it from functools module. 

````python
from functools import reduce

numbers = [1, 2, 3, 4]

def get_product(x, y):
    return x * y

# in first iteration, x will be 0th element and y will be 1st element from list and return value will be product of the two. 
# in next iteration, x will be result for previos return value and y will be next itetrable from the list. 
result = reduce(get_product, numbers)
print(result)
````

````python
from functools import reduce

numbers = [1, 2, 3, 4]

# in first iteration, x will be 0th element and y will be 1st element from list and return value will be product of the two. 
# in next iteration, x will be result for previos return value and y will be next itetrable from the list. 
result = reduce(lambda x, y: x * y, numbers)
print(result)
````

````python
# example of using initial value in reduce
from functools import reduce

numbers = [1, 2, 3, 4]

# in first iteration, x will be initial state which is 100 in this case and y will be 0th element from list and return value will be product of the two. 
# in next iteration, x will be result for previos return value and y will be next itetrable from the list. 
result = reduce(lambda x, y: x * y, numbers, 100)
print(result)
````

## ⚡ Topic 4: Generators & The `yield` Keyword (Memory Magic)

In a normal function, when Python hits the `return` keyword, the function spits out the final result and **dies**. All its internal variables are erased from memory.

But what if you are processing a file with 10 Billion rows? If you `return` a list of 10 Billion items, your computer's RAM will crash.

**Generators** solve this. Instead of `return`, they use the keyword `yield`.
`yield` means: *"Here is one piece of data. Now PAUSE the function, freeze my variables, and wait until you ask for the next piece."*

### A generator:

1. produces a value

2. pauses execution

3. resumes when the next value is requested

````python
# 1. A Normal Function (Returns a list all at once)
def normal_function():
    return [1, 2, 3]
````

````python
# 2. A Generator Function (Yields one item at a time)
def generator_function():
    print("  -> Generating first number...")
    yield 1
    
    print("  -> Generating second number...")
    yield 2
    
    print("  -> Generating third number...")
    yield 3
````

### How to use a Generator

A generator is a special type of function that produces values one at a time instead of returning everything at once.

If you just call a generator function, it doesn't run the code inside. It creates a "Generator Object" waiting to be triggered.
To trigger it and pull the data out, we use the built-in `next()` function, or we loop through it with a `for` loop.

````python
# Create the generator object (No print statements will run yet!)
my_gen = generator_function()
print(my_gen) # <generator object generator_function at 0x...>
````

````python
# Trigger it one step at a time using next()
print("Value:", next(my_gen)) # Runs up to the first yield, then PAUSES
print("Value:", next(my_gen)) # Resumes, runs up to the second yield, PAUSES
````

````python
# You can also just loop through the remaining items!
print("--- Starting Loop ---")
for number in my_gen:
    print("Value:", number) # Prints 3, then stops safely when the generator is empty
````

### The Real Power: Infinite Sequences & Large Data
Because generators only hold *one item in memory at a time*, they can generate infinite sequences without ever crashing your computer! This is heavily used in Data Engineering and Web Scraping.

````python
def infinite_counter(start=0):
    current = start
    while True: # An infinite loop that WON'T crash your RAM!
        yield current
        current += 1
````

````python
# Setup the counter
counter = infinite_counter(100)
print("Value:", next(counter))
print("Value:", next(counter))
print("Value:", next(counter))
print("Value:", next(counter))
````

### Understanding Generator Execution
The function remembers where it stopped.

````python
def demo():
    print("Start")
    yield 1
    print("Middle")
    yield 2
    print("End")
````

````python
g = demo()

print(next(g))
print(next(g))
````

### 🎓 The StopIteration Error

* Question: What happens if I call `next()` when the generator is empty?
* Answer: Python raises a `StopIteration` error. Explain that this error isn't a "bad" crash; it's actually a hidden signal. A `for` loop essentially just calls `next()` over and over again, and when it sees the `StopIteration` error, it catches it silently and stops the loop.

````python
def generator_function():
    print("  -> Generating first number...")
    yield 1
    
    print("  -> Generating second number...")
    yield 2
    
    print("  -> Generating third number...")
    yield 3

print("Value:", next(generator_function()))
print("Value:", next(generator_function()))
print("Value:", next(generator_function()))
````

````python
print("Value:", next(generator_function())) # This will raise a StopIteration
````
