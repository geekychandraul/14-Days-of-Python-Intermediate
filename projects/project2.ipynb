## Mini Data Processing Pipeline

### Objective

Build a data processing pipeline in Python that reads messy order data from a CSV file, cleans and transforms the data, and writes the result into a structured JSON file.

The pipeline must be memory efficient and handle invalid data gracefully.

Students should use concepts they have learned:

- file handling
- CSV processing
- JSON writing
- generators
- exception handling
- datetime parsing
- comprehensions or functional tools
- modular functions

### Input Data : 
Create a file called `orders_raw.csv` and add below data.

`````python
order_id,customer,timestamp,product,quantity,price,status
1001,Alice,2026-01-01T10:00:00,Laptop,1,70000,completed
1002,Bob,2026-01-01T10:05:00,Mouse,2,500,completed
1003,Alice,2026-01-01T10:07:00,Keyboard,1,2000,completed
1004,Charlie,2026-01-01T11:00:00,Monitor,2,12000,completed
1005,David,invalid_timestamp,Laptop,1,70000,completed
1006,Eva,2026-01-01T11:30:00,Mouse,invalid_qty,500,completed
1007,Alice,2026-01-01T12:00:00,Mouse,1,500,cancelled
1008,Bob,2026-01-01T12:30:00,Keyboard,2,2000,completed
1002,Bob,2026-01-01T10:05:00,Mouse,2,500,completed
`````

The dataset intentionally contains real-world problems:

- duplicate orders
- invalid timestamps
- invalid quantity values
- cancelled orders

Students must handle them properly.

### Expected Output
File: `orders_clean.json`

Example:
```JSON
{
  "orders": [
    {
      "order_id": 1001,
      "customer": "Alice",
      "product": "Laptop",
      "quantity": 1,
      "price": 70000,
      "total": 70000,
      "timestamp": "2026-01-01T10:00:00"
    },
    {
      "order_id": 1002,
      "customer": "Bob",
      "product": "Mouse",
      "quantity": 2,
      "price": 500,
      "total": 1000,
      "timestamp": "2026-01-01T10:05:00"
    }
  ]
}
```

### Processing Rules
Students must implement the following logic.

#### 1. Remove Invalid Rows
Skip rows if:

- timestamp is invalid
- quantity cannot be converted to integer
- price cannot be converted to number

Log or print an error message.

#### 2. Ignore Cancelled Orders

Rows with:

`status = cancelled`

must not appear in output.

#### 3. Remove Duplicate Orders

Two rows with the same:

`order_id`

should appear only once.

#### 4. Calculate Total

For every valid order compute:

`total = quantity * price`

#### 5. Convert Timestamp

Convert timestamp into **ISO format using datetime**.

### Required Pipeline Architecture
Students must implement the pipeline using generator functions.

```bash
CSV Reader
     ↓
Row Validator
     ↓
Duplicate Filter
     ↓
Data Transformer
     ↓
JSON Writer
```

### Constraints

Students must follow these constraints.

1. The program should not load the entire CSV into memory.

    - Bad example:
        `rows = list(csv.reader(file))`

2. Must use generator function whereever possible.

3. Must use exception handling for invalid data.

4. Must use datetime module to parse timestamps.

5. Must modularize logic into multiple functions.

6. Duplicate detection should be efficient (using set).
