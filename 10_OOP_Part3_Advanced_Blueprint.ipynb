## 🏢 Topic 1: Class Variables
Instance variables (`self.name`) are unique to each employee.
**Class Variables** belong to the class itself. They are shared by *every* instance.

```python
class Employee:
    # 1. CLASS VARIABLES (Defined outside __init__)
    company_name = "TechCorp"
    raise_amount = 1.05  # 5% raise
    num_of_emps = 0      # Tracker
    
    def __init__(self, name, salary):
        self.name = name       # Instance Variable
        self.salary = salary   # Instance Variable
        
        # We use the Class Name to access/modify class variables
        Employee.num_of_emps += 1
        
    def apply_raise(self):
        # We access the class variable to apply the math
        self.salary = int(self.salary * Employee.raise_amount)

print(f"Initial Employees: {Employee.num_of_emps}")

emp1 = Employee("Alice", 50000)
emp2 = Employee("Bob", 60000)

print(f"Total Employees now: {Employee.num_of_emps}")

# Changing a class variable affects EVERYONE
print(f"Before company rename: {emp1.company_name}")
Employee.company_name = "GlobalTech"
print(f"After company rename: {emp2.company_name}")
```

### Class Variables vs. Hardcoding
* Students asked: "Why put `raise_amount` as a class variable instead of just writing `self.salary * 1.05` inside the method?" 
* Answer: "Because if the CEO changes the raise to 6%, you only have to update ONE variable at the top of the file, not search through 1,000 lines of code for every `1.05`."

## 🛠️ Topic 2: Class Methods & Static Methods
If instance methods take `self` (the object), what takes the Blueprint itself?

* **`@classmethod`**: Takes `cls` (the class) instead of `self`. Used to modify class variables or create alternative constructors.
* **`@staticmethod`**: Takes *neither* `self` nor `cls`. It behaves like a normal function but is kept inside the class because it logically belongs there.

```python
class Employee:
    raise_amount = 1.05
    
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    # CLASS METHOD: Modifies the Blueprint
    @classmethod
    def set_raise_amt(cls, amount):
        cls.raise_amount = amount
        print(f"New company-wide raise is {amount}")

    # CLASS METHOD: Alternative Constructor (Parsing a string to build an object)
    @classmethod
    def from_string(cls, emp_str):
        name, salary = emp_str.split('-')
        return cls(name, int(salary)) # Returns a new Employee object!

    # STATIC METHOD: Just a utility function
    @staticmethod
    def is_workday(day):
        if day.weekday() == 5 or day.weekday() == 6: # Saturday/Sunday
            return False
        return True

# 1. Using a Class Method to change company policy
Employee.set_raise_amt(1.10)

# 2. Using a Class Method to build an object from raw data
new_emp = Employee.from_string("Charlie-75000")
print(f"Welcome {new_emp.name}, starting at ${new_emp.salary}")

# 3. Using a Static Method (Notice we don't need an object to call it!)
import datetime
my_date = datetime.date(2026, 2, 22) # Sunday
print(f"Is today a workday? {Employee.is_workday(my_date)}")
```

## 🪄 Topic 3: Magic Methods (Dunders)
Python uses "Double Underscore" methods (Dunders) for its built-in operations.
If you print an object, Python looks for `__str__`.
If you add two objects (`+`), Python looks for `__add__`.

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    # Dunder for beautifully printing the object
    def __str__(self):
        return f"Employee: {self.name} | Salary: ${self.salary}"
        
    # Dunder for developer debugging
    def __repr__(self):
        return f"Employee('{self.name}', {self.salary})"

    # Dunder to define what the '+' sign does!
    def __add__(self, other_emp):
        # Let's say adding two employees gives us their combined budget
        return self.salary + other_emp.salary
    
emp1 = Employee("Alice", 50000)
emp2 = Employee("Bob", 60000)

# 1. Python uses __str__ automatically here!
print(emp1) 

# 2. Python uses __add__ automatically here!
combined_budget = emp1 + emp2
print(f"Combined team budget is: ${combined_budget}")
```

### `__str__` vs `__repr__` (The Pro Tip)
* `__str__` is for the **End User** (it should be readable and pretty). 
* `__repr__` is for the **Developer** (it should show the exact code needed to recreate the object, like `Employee('Alice', 50000)`). If `__str__` is missing, Python falls back to `__repr__`.

### Demystifying Python's Core
Python isn't magical. When you write `4 + 4`, Python is literally just calling `int.__add__()` behind the scenes. When you implement Dunders, you are proving that your custom objects are first-class citizens in Python.

## Operator Overloading (Polymorphism)
* The `+` sign has many forms. It means "Addition" for integers, "Concatenation" for strings, and "Combine Teams" for our custom Department class. Python figures out what to do by looking at the specific class's `__add__` method. This is Polymorphism in its purest form.
