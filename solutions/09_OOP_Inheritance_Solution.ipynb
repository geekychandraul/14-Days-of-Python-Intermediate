```python
# Level 1: Basic Child

class Vehicle:
    def move(self):
        print("Moving...")

class Car(Vehicle):
    pass

c = Car()
c.move()
```

```python
# Level 2: Adding Tricks

class Car(Vehicle): # Redefining for the example
    def honk(self):
        print("Beep Beep!")

c = Car()
c.move()
c.honk()
```

```python
# Level 3: The Rebel (Overriding)

class Shape:
    """Base class for all shapes"""
    def area(self):
        # We raise an error if someone tries to call this on the base class
        raise NotImplementedError("Subclasses must implement the area method")

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

class Square(Rectangle):
    def __init__(self, side):
        # A square is just a rectangle where width and height are the same
        super().__init__(side, side)

# --- Execution ---

rect = Rectangle(10, 5)
sq = Square(4)

print(f"Rectangle Area: {rect.area()}") # Output: 50
print(f"Square Area: {sq.area()}")       # Output: 16
```

```python
# Level 4: Super Constructor

class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, grade):
        super().__init__(name) # Parent handles name
        self.grade = grade     # Child handles grade

std = Student("Peter", "A+")
print(f"Student: {std.name}, Grade: {std.grade}")
```

```python
# Level 5: E-Commerce System

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    
    def get_info(self):
        return f"Product: {self.name}, Price: ${self.price}"

class Electronic(Product):
    def __init__(self, name, price, warranty):
        super().__init__(name, price)
        self.warranty = warranty
    
    def get_info(self):
        # Calls Parent's get_info first, then adds to it
        base_info = super().get_info()
        return f"{base_info}, Warranty: {self.warranty} years"

laptop = Electronic("MacBook", 1200, 2)
print(laptop.get_info())
```
