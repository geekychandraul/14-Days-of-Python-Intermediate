```python
from datetime import datetime, timedelta
from zoneinfo import ZoneInfo # Python 3.9+
from pathlib import Path
import shutil
```

```python
# Level 1: Future Date Calculator
now = datetime.now()
future = now + timedelta(days=30, hours=12)
print("Future:", future)
```

```python
# Level 2: Timezone Aware

tokyo_time = datetime.now(ZoneInfo("Asia/Tokyo"))
ny_time = datetime.now(ZoneInfo("America/New_York"))
print("Tokyo:", tokyo_time)
print("New York:", ny_time)
```

```python
# Level 3 : Modern paths

folder = Path("project_files")
folder.mkdir(exist_ok=True)
file_path = folder / "log.txt"  # Pathlib magic slash!
file_path.touch()
print("File exists?", file_path.exists())
```

```python
# Level 4: The Mover

backup_dir = Path("backup")
backup_dir.mkdir(exist_ok=True)
shutil.copy(file_path, backup_dir / "log.txt")
print("Backup created?", (backup_dir / "log.txt").exists())
```

```python
# Level 5: Old File Cleaner

today = datetime.now()
cutoff = today - timedelta(days=30) # 30 days ago

# Example file date (Old)
file_date = datetime(2020, 1, 1)

if file_date < cutoff:
    print(f"File from {file_date} is too old. DELETE.")
else:
    print("File is new. KEEP.")
```
