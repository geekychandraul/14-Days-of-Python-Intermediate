```python
# Level 1: Lambda Square

get_square = lambda n: n * n
print(get_square(8)) # 64
```

```python
# Level 2: Name Filter

names = ["Alice", "Bob", "Amanda", "Charlie", "Anna"]
a_names = list(filter(lambda x: x.startswith("A"), names))
print(a_names)
```

```python
# Level 3: Temperature Converter

celsius = [0, 20, 30, 100]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)
```

```python
# Level 4: Flexible Multiplier (*args)

def multiply_all(*args):
    result = 1
    for num in args:
        result *= num
    return result

print(multiply_all(2, 3, 4))
```

```python
# Level 5: HTML Tag Generator (**kwargs)

def make_tag(text, **attributes):
    # Start the tag
    tag_str = "<a"
    
    # Add attributes
    for key, value in attributes.items():
        tag_str += f' {key}="{value}"'
        
    # Close the tag
    tag_str += f">{text}</a>"
    return tag_str

print(make_tag("Click Me", href="google.com", style="color:red"))
# Output: <a href="google.com" style="color:red">Click Me</a>
```
