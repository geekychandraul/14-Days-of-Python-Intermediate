```python
# Level 1: Lambda Square

get_square = lambda n: n * n
print(get_square(8)) # 64
```

```python
# Level 2: Name Filter

names = ["Alice", "Bob", "Amanda", "Charlie", "Anna"]
a_names = list(filter(lambda x: x.startswith("A"), names))
print(a_names)
```

```python
# Level 3: Temperature Converter

celsius = [0, 20, 30, 100]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)
```

```python
# Level 4: Flexible Multiplier (*args)
from functools import reduce

numbers = [1, 2, 3, 4, 5]
product = reduce(lambda x, y: x * y, numbers)
print("Product:", product)
```

```python
# Level 5: HTML Tag Generator (**kwargs)

from functools import reduce
response_times = [120, 95, 340, 180, 60, 410, 220]

def generate_response_times(times):
    for t in times:
        yield t

# for t in generate_response_times(response_times):
#     print(t)

gen = generate_response_times(response_times)

slow_requests = list(filter(lambda t: t > 200, gen))
print(list(slow_requests))

response_seconds = list(map(lambda t: t / 1000, generate_response_times(response_times)))
print(response_seconds)

average_time = reduce(lambda a, b: a + b, response_seconds) / len(response_seconds)
print("Average slow response time:", round(average_time, 2), "seconds")
```
