```python
# Level 1: Timestamp Generator

import datetime as dt
now = dt.datetime.now()
filename = now.strftime("Log_%Y-%m-%d.txt")
print(f"Generated filename: {filename}")
```

```python
# Level 2: Deadline Calculator

days = int(input("Days until deadline: "))
deadline = dt.datetime.now() + dt.timedelta(days=days)
print(f"Due date: {deadline.strftime('%A, %B %d')}")
```

```python
# Level 3: Folder Maker

import os
folder_name = dt.datetime.now().strftime("%Y-%m-%d")

if not os.path.exists(folder_name):
    os.mkdir(folder_name)
    print(f"Created folder: {folder_name}")
else:
    print("Folder already exists.")
```

```python
# Level 4: File Organizer

files = ["report.pdf", "holiday.jpg", "data.csv", "me.jpg"]

for f in files:
    if f.endswith(".jpg"):
        print(f"Moving {f} -> Images/")
    elif f.endswith(".pdf"):
        print(f"Moving {f} -> Documents/")
    else:
        print(f"Skipping {f}")
```

```python
# Level 5: Old File Cleaner

today = dt.datetime.now()
cutoff = today - dt.timedelta(days=30) # 30 days ago

# Example file date (Old)
file_date = dt.datetime(2020, 1, 1)

if file_date < cutoff:
    print(f"File from {file_date} is too old. DELETE.")
else:
    print("File is new. KEEP.")
```
