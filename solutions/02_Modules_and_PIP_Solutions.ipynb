```python
# Level 5: The "New Project" Setup

import os

project = "My_Data_App"

if not os.path.exists(project):
    os.mkdir(project)
    
    # Create subfolder path: My_Data_App/data
    data_folder = os.path.join(project, "data")
    os.mkdir(data_folder)
    
    # Create main.py
    main_file = os.path.join(project, "main.py")
    with open(main_file, "w") as f:
        f.write("# Main App Code")
        
    print("Project Structure Created!")
else:
    print("Project already exists!")
```
