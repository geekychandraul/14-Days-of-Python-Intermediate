```python
# Level 1: Diary Entry

entry = input("How was your day? ")
with open("resources/diary.txt", "w") as file:
    file.write(entry)
print("Entry saved.")
```

```python
# Level 2: The Reader

with open("resources/diary.txt", "r") as file:
    content = file.read()
print(f"Dear Diary, today you said: {content}")
```

```python
# Level 3: Guestbook

while True:
    name = input("Sign guestbook (or 'exit'): ")
    if name == "exit":
        break
    
    with open("resources/guestbook.txt", "a") as file:
        file.write(name + "\n") # Don't forget the \n or it will be one long line!
```

```python
# Level 4: Line Counter

# Make sure story.txt exists first!
try:
    with open("resources/story.txt", "r") as file:
        lines = file.readlines()
        print(f"The story has {len(lines)} lines.")
except FileNotFoundError:
    print("Please create story.txt first!")
```

```python
# Level 5: Error Logger

logs = ["INFO: System OK", "ERROR: Database Fail", "INFO: User login", "ERROR: WiFi down"]

with open("resources/errors_only.txt", "w") as file:
    for log in logs:
        if "ERROR" in log:
            file.write(log + "\n")

# Verify
with open("errors_only.txt", "r") as f:
    print(f.read())
```
