```python
import json
from pathlib import Path
import csv
```

```python
# Level 1
data_string = '{"name": "Alice", "role": "Admin"}'
data_dict = json.loads(data_string)
data_dict["active"] = True
print(json.dumps(data_dict, indent=4))
```

```python
# Level 2: The Basic CSV Writer

rows = [["Name", "Age"], ["Bob", 25], ["Charlie", 30]]
with open(Path("resources") / "users.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(rows[0])    # Write header
    writer.writerows(rows[1:])  # Write data
```

```python
# Level 3: The DictReader
file_path =Path("resources") / "users.csv"
with file_path.open("r", newline="") as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(f"{row['Name']} is {row['Age']} years old.")
```

```python
# Level 4: JSON File Saver
file_path = Path("resources") / "config.json"
config = {"theme": "dark", "notifications": False}
with file_path.open("w") as f:
    json.dump(config, f)

with file_path.open("r") as f:
    loaded_config = json.load(f)
    print("Loaded:", loaded_config)
```

```python
# Level 5: The Converter (CSV -> JSON)
dir_path =Path("resources")
studnet_file_path = dir_path / "students.csv"
updated_student_file_path = dir_path / "updated_students.csv"
updated_students = []
# Write to New CSV
with studnet_file_path.open("r", newline="") as f_read, updated_student_file_path.open("w", newline="")  as f_write:
    reader = csv.DictReader(f_read)
    fieldnames = reader.fieldnames
    for row in reader:
        row["Age"] = int(row["Age"]) + 10
        updated_students.append(row)
        fieldnames = list(row.keys())
    writer = csv.DictWriter(f_write, fieldnames=list(updated_students[0].keys()))
    writer.writeheader()
    writer.writerows(updated_students)

json_file_path = dir_path / "updated_studnets.json"
with json_file_path.open("w") as f:
    json.dump(updated_students, f, indent=4)
print("Migration Complete!")
```
