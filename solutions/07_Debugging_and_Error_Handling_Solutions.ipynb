```python
# Level 1: Safe Calculator

def safe_square(n):
    try:
        num = int(n)
        res = num ** 2
    except ValueError:
        print(f"‚ùå '{n}' is not a number!")
    else:
        print(f"‚úÖ Square is {res}")
```

```python

# Level 3: Custom Password

class WeakPasswordError(Exception): 
    pass

def register(pwd):
    if len(pwd) < 6: 
        raise WeakPasswordError("Too short!")
    print("Password OK.")

try:
    register("123")
except WeakPasswordError as e:
    print(f"Security Alert: {e}")
```

```python
# Level 4: Login System

class UserNotFoundError(Exception): 
    pass
class WrongPasswordError(Exception): 
    pass
db = {"admin": "1234"}

def login(u, p):
    if u not in db:
        raise UserNotFoundError("User missing.")
    if db[u] != p:
        raise WrongPasswordError("Bad password.")
    print("Welcome!")

try:
    login("admin", "wrong")
except UserNotFoundError as e:
    print(f"Login Failed: {e}")
except WrongPasswordError as e:
    print(f"Login Failed: {e}")
```

```python
# Level 5: Reliable Server

class ConnectionError(Exception): pass

def connect(attempt):
    if attempt < 3:
        raise ConnectionError("Server Busy")
    return "Success"

for i in range(1, 6):
    print(f"\n--- Attempt {i} ---")
    try:
        status = connect(i)
    except ConnectionError as e:
        print(f"‚ö†Ô∏è Failed: {e}")
    else:
        print(f"‚úÖ {status}!")
        break
    finally:
        print("Cleanup done.")
```
