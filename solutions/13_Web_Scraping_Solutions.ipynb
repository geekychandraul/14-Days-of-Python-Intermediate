```python
import requests
from bs4 import BeautifulSoup
```

```python
# Practice 1: Cooking the Soup

res = requests.get("http://books.toscrape.com/")
soup = BeautifulSoup(res.text, "html.parser")
print("Title:", soup.title.text.strip())
```

```python
# Practice 2: Find the Author

url = "http://quotes.toscrape.com/"
soup = BeautifulSoup(requests.get(url).text, "html.parser")
first_author = soup.find("small", class_="author")
print("First Author:", first_author.text)
```

```python
# Practice 3: Loop the Authors

authors = soup.find_all("small", class_="author")
for author in authors[:5]:
    print(author.text)
```

```python
# Practice 4: Grab the Login Link

login_tag = soup.find("a", href="/login")
print("Login Link extracted:", login_tag['href'])
```

```python
url = "http://quotes.toscrape.com/"
soup = BeautifulSoup(requests.get(url).text, "html.parser")

# Level 1: Tag Collector

tags = soup.find_all("a", class_="tag")
for tag in tags[:5]:
    print("Tag:", tag.text)
```

```python
# Level 2: Parallel Lists

quotes_list = [q.text for q in soup.find_all("span", class_="text")]
authors_list = [a.text for a in soup.find_all("small", class_="author")]
print(f"Match check: {authors_list[4]} said {quotes_list[4]}")
```

```python
# Level 3: Dictionary Builder

scraped_data = []
# zip() is a great Python built-in to loop through two lists at once!
for q, a in zip(quotes_list, authors_list):
    scraped_data.append({"quote": q, "author": a})

print("\nLevel 3 Output:")
print(scraped_data[:2])
```

```python
# Level 4: Biography Linker
print("\nLevel 4 Output:")
# Find all links where the href starts with "/author/"
bio_links = soup.find_all("a", href=True)
count = 0
for link in bio_links:
    if "/author/" in link['href']:
        full_url = "http://quotes.toscrape.com" + link['href']
        print("Bio Link:", full_url)
        count += 1
        if count == 3: 
            break
```

```python
# Level 5: OOP Scraper

import csv
class QuoteScraper:
    def __init__(self, target_url):
        self.url = target_url
        self.data = []
        
    def fetch_data(self):
        res = requests.get(self.url)
        soup = BeautifulSoup(res.text, "html.parser")
        
        quotes = soup.find_all("span", class_="text")
        authors = soup.find_all("small", class_="author")
        
        for q, a in zip(quotes, authors):
            self.data.append({"quote": q.text, "author": a.text})
        print(f"✅ Fetched {len(self.data)} quotes.")
            
    def save_to_csv(self, filename):
        if not self.data:
            print("No data to save!")
            return
            
        with open(filename, "w", newline="", encoding="utf-8") as f:
            writer = csv.DictWriter(f, fieldnames=["quote", "author"])
            writer.writeheader()
            writer.writerows(self.data)
        print(f"✅ Saved to {filename}.")

# Test
scraper = QuoteScraper("http://quotes.toscrape.com/")
scraper.fetch_data()
scraper.save_to_csv("quotes.csv")
```
