```python
import requests

# Level 1: Dog Fetcher

res = requests.get("https://dog.ceo/api/breeds/image/random")
data = res.json()
print("Dog Image URL:", data["message"])
```

```python
# Level 2: Error Handler

try:
    res = requests.get("https://this-website-is-fake-123.com")
except requests.exceptions.RequestException as e:
    print("Failed to connect to the internet!")
```

```python
# Level 3: User Directory

res = requests.get("https://jsonplaceholder.typicode.com/users")
users = res.json()
for user in users:
    print(f"{user['name']} | {user['email']}")
```

```python
# Level 4: Country Fact Finder

base_url = "https://restcountries.com/v3.1/name/"
# user_input = input("Enter a country name: ") # Uncomment to use!
user_input = "brazil" 
res = requests.get(base_url + user_input)

if res.status_code == 200:
    data = res.json()
    country_data = data[0] # Grab first result
    print(f"Region: {country_data['region']}")
    print(f"Population: {country_data['population']}")
else:
    print("Country not found.")
```

```python
# Level 5: OOP API Client

class PokemonAPI:
    def __init__(self):
        self.base_url = "https://pokeapi.co/api/v2/pokemon/"
        
    def get_weight(self, name):
        name = name.lower() # Safety check
        res = requests.get(self.base_url + name)
        
        if res.status_code == 200:
            data = res.json()
            return f"{name.title()} weighs {data['weight']} hectograms."
        elif res.status_code == 404:
            return f"❌ {name.title()} not found!"
        else:
            return "API Error."

# Test
pokedex = PokemonAPI()
print(pokedex.get_weight("snorlax"))
print(pokedex.get_weight("batman")) # Will trigger 404 logic
```
