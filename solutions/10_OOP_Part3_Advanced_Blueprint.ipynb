```python
# Level 1: Company Motto

class Employee:
    company_motto = "Work Hard, Code Harder"

print(Employee.company_motto)
```

```python
# Level 2: The Re-brander

class Employee:
    company_motto = "Work Hard, Code Harder"
    
    @classmethod
    def change_motto(cls, new_motto):
        cls.company_motto = new_motto

Employee.change_motto("Innovate Daily")
print(Employee.company_motto)
```

```python
# Level 3: The Tax Calculator

class Employee:
    @staticmethod
    def calculate_tax(salary):
        return salary * 0.20

tax = Employee.calculate_tax(100000)
print(f"Tax to pay: ${tax}")
```

```python
# Level 4: The ID Badge

class Employee:
    def __init__(self, name, department):
        self.name = name
        self.department = department
        
    def __str__(self):
        return f"{self.name} - Department of {self.department}"

emp = Employee("Tony", "Engineering")
print(emp)
```

```python

# Level 5: The Department Merger

class Department:
    def __init__(self, name, employees):
        self.name = name
        self.employees = employees
        
    def __len__(self):
        return len(self.employees)
        
    def __str__(self):
        return f"{self.name} Dept with {len(self)} employees"
        
    def __add__(self, other_dept):
        new_name = f"{self.name} & {other_dept.name} Merged"
        new_employees = self.employees + other_dept.employees
        return Department(new_name, new_employees)

# Test Level 5
it_dept = Department("IT", ["Alice", "Bob"])
hr_dept = Department("HR", ["Charlie", "Diana", "Eve"])

merged_dept = it_dept + hr_dept # Calls __add__!

print(merged_dept)       # Calls __str__! (IT & HR Merged Dept with 5 employees)
print(len(merged_dept))  # Calls __len__! (5)
```
