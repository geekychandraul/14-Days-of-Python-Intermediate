```python
import os
```

```python
# Level 1: Inventory Maker

import csv
inventory = [["Item", "Count"], ["Apple", 10], ["Banana", 5]]

with open(os.path.join("resources","inventory.csv"), "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(inventory) # .writerows() writes the whole list at once!
print("Inventory saved.")
```

```python
# Level 2: Price Checker

with open(os.path.join("resources","inventory.csv"), "r", newline="") as file:
    reader = csv.reader(file)
    header = next(reader) # Skips the first row ["Item", "Count"]
    
    for row in reader:
        # row is ["Apple", "10"]
        print(f"We have {row[1]} of {row[0]}")
```

```python
# Level 3: User Profile

import json
my_profile = {
    "name": "Stark",
    "skills": ["Python", "Engineering", "Flying"]
}

with open(os.path.join("resources","profile.json"), "w", newline="") as file:
    json.dump(my_profile, file, indent=4)
print("Profile saved.")
```

```python
# Level 4: Settings Loader

with open(os.path.join("resources","school_config.json"), "r", newline="") as file:
    config = json.load(file)

if config["is_open"]:
    print("School is open!")
else:
    print("School is closed.")
```

```python
# Level 5: The Converter (CSV -> JSON)

# Step 1: Read CSV
students_list = []
with open(os.path.join("resources","students.csv"), "r", newline="") as file:
    reader = csv.reader(file)
    header = next(reader) # ["Name", "Age", "Grade"]
    
    for row in reader:
        # Create a dictionary for this row
        student_dict = {
            "Name": row[0],
            "Age": row[1],
            "Grade": row[2]
        }
        students_list.append(student_dict)

# Step 2: Write JSON
with open(os.path.join("resources","students.json"), "w", newline="") as file:
    json.dump(students_list, file, indent=4)

print("Conversion Complete! Check students.json")
```
