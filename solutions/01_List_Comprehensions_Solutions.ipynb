```python
#Level 1: Number Doubler

numbers = [1, 2, 3, 4, 5]
doubled = [num * 2 for num in numbers]
print(doubled)
```

```python
# Level 2: Name Cleaner

names = ["  alice ", "  bob", "charlie  "]
clean_names = [name.strip().title() for name in names]
print(clean_names)
```

```python
# Level 3: The Filter

prices = [5, 12, 3, 50, 8, 20]
expensive = [p for p in prices if p > 10]
print(expensive)
```

```python
# Level 4: Length Mapper

sentence = "Python is awesome"
# sentence.split() creates ["Python", "is", "awesome"]
word_lengths = {word: len(word) for word in sentence.split()}
print(word_lengths)
```

```python
# Level 5: Matrix Flattener

matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
# Read logic: for row in matrix -> for num in row -> keep num
flat = [num for row in matrix for num in row]
print(flat)
```
