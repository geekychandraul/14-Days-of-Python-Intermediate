```python
# The Parent Class 
class Employee:
    def __init__(self, name, department, salary):
        self.name = name
        self.department = department
        self.__salary = salary # Private
        
    @property
    def salary(self):
        return self.__salary
    
    @salary.setter
    def salary(self, new_salary):
        self.salary = new_salary

    def get_details(self):
        return f"{self.name} works in {self.department}."

class Developer(Employee):
    def write_code(self):
        return f"{self.name} is typing Python code... 💻"
    
    def hike(self):
        self.salary += self.salary * 1.10
        
    def get_details(self):                                                              #1
        return f"{self.name} writes code in Python."
    
developer = Developer("Alice", "Engineering", 90000)
print(developer.salary)
```

```python
# Calling Super outside the class

class Developer(Employee):
    def __init__(self, name, department, salary, programming_language):
        # 1. Let the Parent (Employee) handle the basic setup
        super().__init__(name, department, salary)
        
        # 2. Add the specific Developer attribute
        self.programming_language = programming_language

    def get_details(self):
        # using super to call parent's method
        return super().get_details() + f" They code in {self.programming_language}."    
      
    def write_code(self):
        print(f"{self.name} is typing {self.programming_language} code... 💻")

dev1 = Developer("Alice", "Engineering", 90000, "Python")
print(dev1.super().get_details()) # type: ignore                             #1
```

```python
# Accessing attributes using super()

class Developer(Employee):
    def __init__(self, name, department, salary, programming_language):
        # 1. Let the Parent (Employee) handle the basic setup
        super().__init__(name, department, salary)
        
        # 2. Add the specific Developer attribute
        self.programming_language = programming_language

    def get_details(self):
        # using super to call parent's method
        return super().get_details() + f" They code in {self.programming_language}."    
      
    def get_department(self):
        return super().department                                         # 1


dev1 = Developer("Alice", "Engineering", 90000, "Python")
print(dev1.get_department())
```

### Types of Inheritance

```python
# Single Inheritance
class Developer(Employee):
    def __init__(self, name, department, salary, programming_language):
        # 1. Let the Parent (Employee) handle the basic setup
        super().__init__(name, department, salary)
        
        # 2. Add the specific Developer attribute
        self.programming_language = programming_language

    def get_details(self):
        # using super to call parent's method
        return super().get_details() + f" They code in {self.programming_language}."    
      
    def write_code(self):
        print(f"{self.name} is typing {self.programming_language} code... 💻")

dev1 = Developer("Alice", "Engineering", 90000, "Python")
print(dev1.get_details())
print(dev1.write_code())
```

```python
# Multilevel Inheritance

class Employee:
    
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    
    def show_employee(self):
        print(f"Employee: {self.name}, Salary: {self.salary}")


class Manager(Employee):
    
    def __init__(self, name, salary, team_size):
        super().__init__(name, salary)
        self.team_size = team_size
    
    def show_manager(self):
        print(f"{self.name} manages {self.team_size} people")


class Director(Manager):
    
    def __init__(self, name, salary, team_size, department):
        super().__init__(name, salary, team_size)
        self.department = department
    
    def show_director(self):
        print(f"{self.name} heads the {self.department} department")


d = Director("Anita", 250000, 10, "Engineering")

d.show_employee()
d.show_manager()
d.show_director()
```

```python
# Hierarchical Inheritance

class Employee:
    
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    
    def get_details(self):
        print(f"{self.name} works for the company")


class Developer(Employee):
    
    def code(self):
        print(f"{self.name} writes code")


class Tester(Employee):
    
    def test(self):
        print(f"{self.name} tests software")


class Manager(Employee):
    
    def manage(self):
        print(f"{self.name} manages the team")


dev = Developer("Ravi", 120000)
tester = Tester("Sneha", 100000)
mgr = Manager("Arjun", 150000)

dev.get_details()
dev.code()

tester.get_details()
tester.test()

mgr.get_details()
mgr.manage()
```

```python
# Multiple Inheritance(Diamond Problem)

class Employee:
    
    def work(self):
        print("Employee works")


class Manager(Employee):
    
    def manage(self):
        print("Manager manages team")


class TechLead(Employee):
    
    def design_system(self):
        print("TechLead designs system")


class TechManager(Manager, TechLead):
    
    def role(self):
        print("TechManager handles both management and technical leadership")


tm = TechManager()

tm.work()
tm.manage()
tm.design_system()
tm.role()
```

```python
# Multilevel Inheritance

class Animal:

    def eat(self):
        print("Animal eats")


class Dog(Animal):

    def bark(self):
        print("Dog barks")


class Puppy(Dog):

    def cry(self):
        print("Puppy cries")


p = Puppy()

p.eat()
p.bark()
p.cry()
```

```python
# Hierarchical Inheritance
class Shape:

    def draw(self):
        print("Drawing shape")


class Circle(Shape):

    def area(self):
        print("Area of circle")


class Rectangle(Shape):

    def area(self):
        print("Area of rectangle")


c = Circle()
r = Rectangle()

c.draw()
r.draw()
```

```python
# Multiple Inheritance
class Camera:

    def take_photo(self):
        print("Photo taken")


class Phone:

    def call(self):
        print("Calling...")


class SmartPhone(Camera, Phone):
    pass


s = SmartPhone()

s.take_photo()
s.call()
```

```python
# Access Parent Method Directly
class Vehicle:

    def start(self):
        print("Vehicle starting")


class Car(Vehicle):

    def start(self):
        print("Car starting")
        super().start()


c = Car()
c.start()
```

```python
# Diamond Problem
# Classic multiple inheritance problem.
# https://www.geeksforgeeks.org/python/diamond-problem-in-python/
class A:
    def show(self):
        print("A")

class B(A):
    print("B")


class C(A):
    print("C")


class D(B, C):
    print("D")


d = D()
d.show()
```
