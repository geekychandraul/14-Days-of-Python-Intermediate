```python
import matplotlib.pyplot as plt

class Point:
    def __init__(self, x, y):
        self.x_cod = x
        self.y_cod = y


p1 = Point(1,2)
p2 = Point(3,4)

origin = (0,0)

# Plot the points
plt.scatter([p1.x_cod, p2.x_cod], [p1.y_cod, p2.y_cod])

# Draw line from origin to p1
plt.plot([origin[0], p1.x_cod], [origin[1], p1.y_cod])

# Draw line from origin to p2
plt.plot([origin[0], p2.x_cod], [origin[1], p2.y_cod])

plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Vectors from Origin")

plt.scatter(0,0)  # plot origin

plt.show()
```

```python
class Point:
    def __init__(self,x,y):
        self.x_cod = x
        self.y_cod = y

    def __str__(self):
        return '<{},{}>'.format(self.x_cod,self.y_cod)


    def distance_from_origin(self):
        # Formula : sqrt(x^2 + y^2)   
        return (self.x_cod**2 + self.y_cod**2)**0.5
    
    def euclidean_distance(self,other):
        # Formula: sqrt((x1-x2)^2 + (y1-y2)^2)
        return ((self.x_cod - other.x_cod)**2 + (self.y_cod - other.y_cod)**2)**0.5
    
    def cosine_similarity(self,other):
        # Formula: (x1*x2 + y1*y2)/(||x1,y1||*||x2,y2||)
        # Mod: sqrt(x^2 + y^2)
        return (self.x_cod * other.x_cod + self.y_cod * other.y_cod) / (self.distance_from_origin() * other.distance_from_origin())
    


p1 = Point(1,2)
p2 = Point(3,4)
print(p1)
```

```python

print(p1)
print(p1.distance_from_origin())
print(p1.euclidean_distance(p2))
print(p1.cosine_similarity(p2))
```
