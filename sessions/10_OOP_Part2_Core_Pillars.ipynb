## 🦸 Topic 5: The `super()` Function

### Super to call Parent function

### Parent Child Constructor 

What if our `Developer` needs an extra attribute, like `programming_language`?
If we write a new `__init__` in `Developer`, it will completely override the Parent's `__init__`, and we lose our secure salary setup!

We use `super()` to tell the Parent class to do its normal setup first, and then we add our specific changes

#### Using Method Overloading and Calling Parent's Method

What if we wanted to add more capablilities to Developer's `get_detials` method using the Employee's `get_details` method. 

### Why `super()`?

**The Struggle**: Beginners will try to copy-paste `self.name = name` into the Child's init.

The Fix: *"If the way we store names changes in the Parent class later, your Child class will break. `super()` ensures the Child always follows the Parent's rules."*

````python

class Employee:
    def __init__(self, name, department, salary):
        self.name = name.strip().title()                                                #1
        self.department = department.upper()                                            #2

        if salary < 0:                                                                  #3
            raise ValueError("Salary cannot be negative")
        
        self.__salary = salary

class Developer(Employee):
    def __init__(self, name, department, salary, programming_language):
        self.programming_language = programming_language
        self.name = name                                                                #1
        self.department = department                                                    #2
        self.__salary = salary                                                          #3

emp = Employee(" alice ", "eng", -100)

dev = Developer(" alice ", "eng", -100, "Python")
````

### Edge Cases for Super

#### Calling Super outside the class

````python
# The Parent Class 
class Employee:
    def __init__(self, name, department, salary):
        self.name = name
        self.department = department
        self.__salary = salary # Private
        
    @property
    def salary(self):
        return self.__salary

    def get_salary(self):
        return self.__salary 

    def update_salary(self, new_salary):
        if new_salary < 0:
            raise ValueError("Salary cannot be negative!")
        self.__salary = new_salary   
                
    def get_details(self):
        return f"{self.name} works in {self.department}."
````

````python
class Developer(Employee):
    def __init__(self, name, department, salary, programming_language):
        # 1. Let the Parent (Employee) handle the basic setup
        super().__init__(name, department, salary)

        # 2. Add the specific Developer attribute
        self.programming_language = programming_language

    def get_details(self):
        # using super to call parent's method
        return super().get_details() + f" They code in {self.programming_language}."

    def write_code(self):
        print(f"{self.name} is typing {self.programming_language} code... 💻")


dev1 = Developer("Alice", "Engineering", 90000, "Python")
print(dev1.super().get_details())  # type: ignore                                                                #1
````

#### Accessing attributes using super()

````python
class Developer(Employee):
    def __init__(self, name, department, salary, programming_language):
        # 1. Let the Parent (Employee) handle the basic setup
        super().__init__(name, department, salary)
        self.name = f"MR. {name}"
        # 2. Add the specific Developer attribute
        self.programming_language = programming_language

    def get_details(self):
        # using super to call parent's method
        return super().get_details() + f" They code in {self.programming_language}."    
      
    def get_department(self):
        return super().department                                                                                   #2


dev1 = Developer("Alice", "Engineering", 90000, "Python")
print(dev1.get_details())
````

### Single Inheritance
Single inheritance means one child class inherits from one parent class. The child class automatically gets all accessible attributes and methods of the parent.

```bash
Parent
   ↓
Child
```

````python
class Developer(Employee):
    def __init__(self, name, department, salary, programming_language):
        # 1. Let the Parent (Employee) handle the basic setup
        super().__init__(name, department, salary)
        
        # 2. Add the specific Developer attribute
        self.programming_language = programming_language

    def get_details(self):
        # using super to call parent's method
        return super().get_details() + f" They code in {self.programming_language}."    
      
    def write_code(self):
        print(f"{self.name} is typing {self.programming_language} code... 💻")

dev1 = Developer("Alice", "Engineering", 90000, "Python")
print(dev1.get_details())
print(dev1.write_code())
````

### Multilevel Inheritance

Multilevel inheritance means a class inherits from a class that already inherited from another class. Each level can add new behavior. A class can be parent and can be child also. 

```bash
Grandparent
   ↓
Parent
   ↓
Child
```

````python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    
    def show_employee(self):
        print(f"Employee: {self.name}, Salary: {self.salary}")


class Manager(Employee):
    def __init__(self, name, salary, team_size):
        super().__init__(name, salary)
        self.team_size = team_size
    
    def show_manager(self):
        print(f"{self.name} manages {self.team_size} people")


class Director(Manager):
    def __init__(self, name, salary, team_size, department):
        super().__init__(name, salary, team_size)
        self.department = department
    
    def show_director(self):
        print(f"{self.name} heads the {self.department} department")



d = Director("Anita", 250000, 10, "Engineering")

d.show_employee()
d.show_manager()
d.show_director()
````

### Hierarchical Inheritance

Hierarchical inheritance means multiple child classes inherit from the same parent class. Different child classes share common behavior from the same parent.

```bash
        Employee
      /     |    \
Developer Tester Manager
```

````python
class Employee:
    
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    
    def get_details(self):
        print(f"{self.name} works for the company")


class Developer(Employee):
    
    def code(self):
        print(f"{self.name} writes code")


class Tester(Employee):
    
    def test(self):
        print(f"{self.name} tests software")


class Manager(Employee):
    
    def manage(self):
        print(f"{self.name} manages the team")


dev = Developer("Ravi", 120000)
tester = Tester("Sneha", 100000)
mgr = Manager("Arjun", 150000)

dev.get_details()
dev.code()

tester.get_details()
tester.test()

mgr.get_details()
mgr.manage()
````

### Multiple Inheritance(Diamond Problem)
Multiple inheritance means a class inherits from more than one parent class. The child class receives behavior from both parent classes.

```bash 

        Employee
        /     \
   Manager   TechLead
        \     /
      TechManager

```

````python
class Employee:
    def work(self):
        print("Employee works")


class Manager(Employee):
    def manage(self):
        print("Manager manages team")
        

class TechLead(Employee):
    def design_system(self):
        print("TechLead designs system")
    

class TechManager(Manager, TechLead):
    def role(self):
        print("TechManager handles both management and technical leadership")

tm = TechManager()

tm.work()
tm.manage()
tm.design_system()
tm.role()
````

### Hybrid Inheritance

Hybrid inheritance is a combination of two or more inheritance types. Usually it combines:
- Hierarchical inheritance
- Multiple inheritance

```Bash
          A
        /   \
       B     C
        \   /
          D
```

````python
# Single Inheritance

class Employee:
    def work(self):
        print("Employee works")


class Manager(Employee):
    def manage(self):
        print("Manager manages team")
    
    def work(self):
        print("Manager works")

````

Python follows a specific order to search methods → that order is called MRO.

## 👻 Topic 5: Abstraction (Enforcing Rules)
Imagine you are the CEO. You decree: *"Every employee must have a method to calculate their end-of-year bonus, but the math is different for every role!"*

You can make `Employee` an **Abstract Base Class (ABC)**. 
It forces all Child classes (Developer, Manager) to write a `calculate_bonus()` method. If they don't, Python crashes.

````python
# The Parent Class 
class Employee:
    def __init__(self, name, department, salary):
        self.name = name
        self.department = department
        self.__salary = salary # Private
        
    @property
    def salary(self):
        return self.__salary
    
    @salary.setter
    def salary(self, new_salary):
        if new_salary < 0:
            raise ValueError("Salary cannot be negative!")
        self.__salary = new_salary
               
    def get_details(self):
        return f"{self.name} works in {self.department}."
````

````python
# Convert base employee class into abstract class
from abc import ABC, abstractmethod

# 1. Inherit from ABC to make it abstract
class AbstractEmployee(ABC):
    def __init__(self, name, department, salary):
        self.name = name
        self.department = department
        self.__salary = salary
        
    @property
    def salary(self):
        return self.__salary
    
    @salary.setter
    def salary(self, new_salary):
        if new_salary < 0:
            raise ValueError("Salary cannot be negative!")
        self.__salary = new_salary
        
    # 2. Use the decorator to create a strict rule
    @abstractmethod
    def get_details(self):
        pass # No logic here, just the rule!
````

````python
# create instance of abstract class
emp = AbstractEmployee("Ghost", "Engineering", 0)  # type: ignore
````

````python
# define developer class using inheritance

class Developer(AbstractEmployee):
    def __init__(self, name, department, salary, programming_language):
        super().__init__(name, salary, department)
        self.programming_language = programming_language
````

````python
# get_details method to be required method
from abc import ABC, abstractmethod

class AbstractEmployee(ABC):
    def __init__(self, name, department, salary):
        self.name = name
        self.department = department
        self.__salary = salary
        
    @property
    def salary(self):
        return self.__salary
    
    @salary.setter
    def salary(self, new_salary):
        if new_salary < 0:
            raise ValueError("Salary cannot be negative!")
        self.__salary = new_salary
        
    # 2. Use the decorator to create a strict rule                                                      #1
    @abstractmethod                                                                                     #2
    def get_details(self):
        pass # No logic here, just the rule!
````

````python
# Developer class without abstract method
class Developer(AbstractEmployee):
    def __init__(self, name, department, salary, programming_language):
        super().__init__(name, salary, department)
        self.programming_language = programming_language
    
dev = Developer("Alice", "Engineering", 50000, "Python") # type: ignore
````

````python
# Developer class with abstract method
class Developer(AbstractEmployee):
    def __init__(self, name, department, salary, programming_language):
        super().__init__(name, salary, department)
        self.programming_language = programming_language
    
    def get_details(self):
        return f"{super().get_details()} {self.name} writes code in {self.programming_language}."
````

````python
# Add more capabilities to Developer class
class Developer(AbstractEmployee):
    def __init__(self, name, department, salary, programming_language):
        super().__init__(name, salary, department)
        self.programming_language = programming_language
    # If we delete this method, Python will crash!
    def get_details(self):
        return f"{self.name} is interning in {self.department} and uses {self.programming_language}."

    def calculate_bonus(self):
        return 500 # Interns get a flat $500 bonus
````

## Topic 5: Polymorphism in Action

In Python, Polymorphism allows different classes to be treated as instances of the same general class through the same interface. The word literally means "many forms"—the same method name (interface) takes on different behaviors depending on which object is calling it.

The best way to see this is through a Payment System. Whether you pay via Credit Card or PayPal, the "action" is the same (`process_payment`), but the internal logic is very different.

````python
class Cat:
    def speak(self): 
        return "Meow"

class Dog:
    def speak(self): 
        return "Woof"

class Robot:
    def speak(self): 
        return "Beep Boop"
````

````python
# This function is polymorphic but uses NO inheritance/overriding
def make_it_talk(entity):
    print(entity.speak())

make_it_talk(Cat())
make_it_talk(Robot())
make_it_talk(Dog())
````

````python
class Employee:
    def __init__(self, name, department, salary):
        self.name = name
        self.department = department
        self.__salary = salary # Private
        
    @property
    def salary(self):
        return self.__salary
        
    def get_details(self):
        return f"{self.name} works in {self.department}."

class Developer(Employee):
    def __init__(self, name, department, salary, programming_language):
        super().__init__(name, department, salary)
        self.programming_language = programming_language

    def get_details(self):
        return super().get_details() + f" They code in {self.programming_language}."  
      
    def write_code(self):
        print(f"{self.name} is typing {self.programming_language} code... 💻")
````

See how same method `get_detials` has different behaviour for two differnet class. 


````python
# This function is polymorphic but uses NO inheritance/overriding
def get_details(entity):
    print(entity.get_details())
emp1 = Employee("Bob", "HR", 60000)
dev1 = Developer("Alice", "Engineering", 90000, "Python")

get_details(emp1)
get_details(dev1)
````

````python
class Payment:
    def process_payment(self, amount):
        raise NotImplementedError("Subclass must implement abstract method")

class CreditCard(Payment):
    def process_payment(self, amount):
        return f"Processing credit card payment of ${amount} (Applying 2% transaction fee)."

class PayPal(Payment):
    def process_payment(self, amount):
        return f"Processing PayPal payment of ${amount} (Redirecting to secure portal)."

class Crypto(Payment):
    def process_payment(self, amount):
        return f"Processing Bitcoin payment of ${amount} (Waiting for blockchain confirmation)."
````

````python
# --- THE POLYMORPHIC FUNCTION ---
def checkout(payment_method, amount):
    # This function doesn't care what 'type' payment_method is.
    # It only cares that it HAS a .process_payment() method.
    print(payment_method.process_payment(amount))

# Using different objects in the same function
methods = [CreditCard(), PayPal(), Crypto()]

for method in methods:
    checkout(method, 100)
````

### Why is this powerful?
**Flexibility**: You can add a new payment method (like ApplePay) next week. You only need to create the class and the process_payment method. You don't have to change the checkout function at all.

**Decoupling**: The checkout logic is separated from the specific details of how each payment works.

**Duck Typing**: Python follows the "Duck Typing" philosophy: "If it walks like a duck and quacks like a duck, it’s a duck." In our example, if an object has a process_payment method, Python treats it as a valid payment method.
