## 🏗️ Topic 1: The Soup (Parsing HTML)
When you visit a website, your browser reads HTML tags like `<h1>` (Header), `<p>` (Paragraph), and `<a>` (Link). 
If we just use `requests.get()`, we get a giant, messy string of HTML. `BeautifulSoup` turns that messy string into a searchable tree.

````python
import requests
from bs4 import BeautifulSoup

# 1. Fetch the raw HTML from our practice website
url = "http://quotes.toscrape.com/"
response = requests.get(url)

# 2. Cook the "Soup" (Parse the HTML)
# We pass the raw text, and tell it to use the built-in HTML parser
soup = BeautifulSoup(response.text, "html.parser")

# Now we can easily search it! Let's grab the title of the webpage.
print("Raw Title Tag:", soup.title)
print("Clean Text Only:", soup.title.text)
````

## 🎯 Topic 2: Finding a Single Element (`.find()`)

If you want to find the first instance of a tag, use `soup.find()`.
You can make your search specific by passing the HTML `class` or `id`.

Note: Because `class` is a reserved keyword in Python (used for OOP), BeautifulSoup uses `class_` (with an underscore).

````python
url = "http://quotes.toscrape.com/"
soup = BeautifulSoup(requests.get(url).text, "html.parser")

# 1. Find the first <h1> tag
main_header = soup.find("h1")
print("Header:", main_header.text.strip()) # .strip() removes extra spaces/newlines

# 2. Find a specific tag by its class
# Looking at the website's code, quotes are in <span class="text">
first_quote = soup.find("span", class_="text")
print("\nFirst Quote:", first_quote.text)
````

## 🗄️ Topic 3: Finding Multiple Elements (`.find_all()`)

Usually, you don't just want one item; you want all the items on the page (like a list of products).
`soup.find_all()` returns a List of tags that you can loop through!

````python
url = "http://quotes.toscrape.com/"
soup = BeautifulSoup(requests.get(url).text, "html.parser")

# Find ALL quotes on the page
all_quotes = soup.find_all("span", class_="text")

print(f"Found {len(all_quotes)} quotes on this page!\n")

# Loop through the list to print the first 3
for quote in all_quotes[:3]:
    print("-", quote.text)
````

## 🔗 Topic 4: Extracting Attributes (Links & Images)


Sometimes the data you want isn't the text between the tags, but the data inside the tag itself.
For example, in `<a href="https://google.com">Click Here</a>`, we want the `href` link, not the text "Click Here".

BeautifulSoup treats tags like Dictionaries. You can access attributes using `tag['href']` or `tag['src']`.

````python
url = "http://quotes.toscrape.com/"
soup = BeautifulSoup(requests.get(url).text, "html.parser")

# 1. Find the "Next Page" button
# The HTML looks like: <li class="next"><a href="/page/2/">Next</a></li>
next_li = soup.find("li", class_="next")

if next_li:
    # Dig inside the <li> to find the <a> tag
    link_tag = next_li.find("a")
    
    # Extract the 'href' attribute like a dictionary!
    partial_link = link_tag["href"] 
    
    print("Found next page link:", partial_link)
    
    # We can reconstruct the full URL!
    full_url = "http://quotes.toscrape.com" + partial_link
    print("Full URL:", full_url)
````
