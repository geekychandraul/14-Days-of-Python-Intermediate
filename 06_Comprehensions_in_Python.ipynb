## 🚀 Topic 1: The Old Way vs. The New Way
Imagine we want a list of squares for numbers 0-9: `[0, 1, 4, 9, ...]`.

**The "Basic" Way (3 Lines):**


```python
squares = []
for i in range(10):
    squares.append(i ** 2)
print(squares)
```

**The "Intermediate" Way (1 Line):**

```python
# [Expression for Item in Iterable]
squares = [i ** 2 for i in range(10)]
print(squares)
```

```python
names = ["tony", "steve", "thor", "bruce"]

# We want: ["T", "S", "T", "B"]
initials = [name[0] for name in names]

print(initials)

# Make initials uppercase too!
print([name.capitalize() for name in names])
```

This is called a List Comprehension. It is faster to write and faster for the computer to read.

#### The "Readability" Argument

**Students asks**: "Why is this better? The for loop was easier to read."

**Answer**: It takes practice. Once you get used to it, reading [x*2 for x in nums] is like reading an English sentence. It focuses on what you want (the result), not how to get it (the loop mechanics).

## 🔍 Topic 2: Adding Logic (Filtering)

What if we only want squares of Even numbers? We can add an if statement at the end.

**Syntax**: [expression for item in list if condition]

```python
# Old Way
evens = []
for i in range(20):
    if i % 2 == 0:
        evens.append(i)

# Pythonic Way
# "Give me i, for every i in range 20, IF i is even"
# Execution goes from left to right and then value is added to the list if the condition is satisfied
evens_pro = [i for i in range(20) if i % 2 == 0]
print(f"Old: {evens}")
print(f"Pro: {evens_pro}")

# Execution goes from left to right and then value is added to the list if the condition is satisfied
evens_pro_tuple = [(i, j) for i in range(20) for j in range(10) if i % 2 == 0 and j % 2 == 0 if i > 10]
print(f"Pro Tuple: {evens_pro_tuple}")
```

### Adding nested loops in List Comprehensions
What if we want all combinations of two lists? We can add another for loop.

**Syntax**: [expression for item in list for item2 in list2 if condition]

```python
# Execution goes from left to right and then value is added to the list if the condition is satisfied
evens_pro_tuple = [(i, j) for i in range(20) for j in range(10) if i % 2 == 0]
print(f"Pro Tuple: {evens_pro_tuple}")

# adding multiple if condition to nested loop
evens_pro_tuple_gt_ten = [(i, j) for i in range(20) for j in range(10) if i % 2 == 0 and j % 2 == 0 if i > 10]
print(f"Pro Tuple: {evens_pro_tuple_gt_ten}")
```

#### The Order of Operations

`[item for sublist in list for item in sublist]`

**Explainaton**: It's like a nested loop

Outer loop (for sublist in list) comes first.

Inner loop (for item in sublist) comes second.

### When NOT to use Comprehensions

**Warning**: If your logic is super complex (more than 2 if statements), do not use a comprehension. Go back to a normal loop. Code is for humans to read first, computers second.

## 🧠 Topic 3: Dictionary Comprehensions
Yes, you can do this with Dictionaries too!
**Syntax:** `{key: value for item in list}`

Just like List Comprehensions, you can add if statements and nested loops to Dictionary Comprehensions.

```python
# Example 1: Swapping Keys and Values
original = {"A": 1, "B": 2}
swapped = {value ** 2: key for key, value in original.items()}
print(swapped) # {1: "A", 2: "B"}


# Intro to zip() method
values = ["Tony", 45, "Ironman"]
keys = ["name", "age", "role"]

# Creating a dict from two lists using zip()
# zip() pairs them up: ("name", "Tony"), ("age", 45)...
# Just like dict.items(), zip also gives us (key, value) pairs from a dict
print(list(zip(keys, values)))

# Dict comprehension to create a dict from two lists
my_dict = {k: v for k, v in zip(keys, values) if isinstance(v, str)}

print(my_dict)
```

## Topic 4: Set Comprehensions
Set Comprehension creates a set, which means:
- unique elements
- unordered collection

**Syntax**: {expression for item in iterable}

In set comprehensions, duplicates disappear automatically.

Just like lists and dictionaries, you can add if statements and nested loops to Generators.

```python
emails = ["a@gmail.com", "b@yahoo.com", "c@gmail.com"]

domains = {email.split("@")[1] for email in emails}
print(domains)
```

## Topic 5: Tuple Comprehensions (Generators)
You can also create tuples with comprehensions, but they are called Generators. They use parentheses instead of square brackets.

**Syntax**: `(expression for item in list)`

Generators are more memory efficient than lists because they generate items on the fly instead of storing them all in memory.

Just like lists and dictionaries, you can add if statements and nested loops to Generators.

```python
# tuple comprehension
numbers = [1, 2, 3, 4, 5]
squares = [num ** 2 for num in numbers] # Memory is allocated
print(squares)

squares_tup = (num ** 2 for num in numbers) # Memory is not allocated
print(squares_tup)
print(next(squares_tup))
print(next(squares_tup))
print(next(squares_tup))
print(next(squares_tup))
print(next(squares_tup))
```

```python
# What if we do next on a generator when all values are already consumed?
print(next(squares_tup))  # StopIteration Exception
```
